```python
import pandas as pd
import sqlite3
import glob
import os
```

```python
eventTypeTableFilenames = [fn for fn in glob.glob('event_*.csv') if not fn == "event_object.csv"]
objectTypeTableFilenames =  [fn for fn in glob.glob('object_*.csv') if not fn == "object_object.csv"]

TABLES = dict()
TABLES["event"] = pd.read_csv("event.csv", sep=";")
TABLES["event_object"] = pd.read_csv("event_object.csv", sep=";")
TABLES["object"] = pd.read_csv("object.csv", sep=";")
TABLES["object_object"] = pd.read_csv("object_object.csv", sep=";")
for fn in eventTypeTableFilenames:
    table_name = fn.split(".")[0]
    TABLES[table_name] = pd.read_csv(fn, sep=";")
for fn in objectTypeTableFilenames:
    table_name = fn.split(".")[0]
    TABLES[table_name] = pd.read_csv(fn, sep=";")
```

```python
sql_paths = ["order_management_0.sqlite"]
for sql_path in sql_paths:
    if os.path.exists(sql_path):
        os.remove(sql_path)
```

```python
for sql_path in sql_paths:
    conn = sqlite3.connect(sql_path)
    for tn, df in TABLES.items():
        df.to_sql(tn, conn, index=False)
    conn.close()
```
